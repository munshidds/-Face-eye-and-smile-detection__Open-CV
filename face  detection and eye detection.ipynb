```python
import cv2
```

```python
image=cv2.imread(r'C:\Users\PRO\Downloads\faces.jpg')
haar_cascade=cv2.CascadeClassifier(r'C:\Users\PRO\Downloads\haarcascade_frontalface_default.xml')
#detctMultiscale(image,scale factor ,min_meighours)
#scale factor : parameter specifying how much the image size is reduced  at each image scale
#min_neighbour:parmeter specifiing how many neighbors 

faces=haar_cascade.detectMultiScale(image,1.1,9)
faces
```

```python
import matplotlib.pyplot as plt
plt.imshow(image)
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y), (x+w,y+h),(0,0,255),2)
cv2.imshow('detected_faces',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

eye detection

```python
image=cv2.imread(r'C:\Users\PRO\Downloads\faces.jpg')
eyes_cascade=cv2.CascadeClassifier(r'C:\Users\PRO\Downloads\haarcascade_eye.xml')
for (x,y,w,h) in faces:
    cv2.rectangle(image,(x,y),(x+w,y+h),(0,0,255),2)
    #rop part of the image containing face
    sub_image=image[y:y+h,x:x+w]
    eyes=eyes_cascade.detectMultiScale(sub_image)
    for (x,y,w,h) in eyes:
        cv2.rectangle(sub_image,(x,y),(x+w,y+h),(0,0,255),2)
cv2.imshow('detected_face_and_eye',image)
cv2.imshow('sub_image',sub_image)


cv2.waitKey(0)
cv2.destroyAllWindows()
```
